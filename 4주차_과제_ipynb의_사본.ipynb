```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick



fig, axes = plt.subplots(3, 2, figsize=(16, 14))
fig.suptitle('일 단위 비율 (FIRM vs FORECAST)', fontsize=13, fontweight='bold')

for ax, pn in zip(axes.flat, PN_LIST):
    g = df[df['ZF_PN'] == pn]

    daily = (g.groupby(['Date', 'Firm/Forecast'])['Quantity']
               .sum().unstack(fill_value=0).sort_index())


    ratio = daily.div(daily.sum(axis=1), axis=0)

    firm_vals = ratio.get('FIRM', pd.Series(dtype=float))
    fc_vals   = ratio.get('FORECAST', pd.Series(dtype=float))

    x = range(len(ratio))

    ax.fill_between(x, 0, fc_vals.values,
                    color='orange', alpha=0.3, label='FORECAST')

    ax.fill_between(x, fc_vals.values, fc_vals.values + firm_vals.values,
                    color='blue', alpha=0.4, label='FIRM')

    step = max(1, len(ratio) // 10)
    tick_pos   = list(range(0, len(ratio), step))
    tick_labels= [str(d.date())[5:] for d in ratio.index[tick_pos]]

    ax.set_xticks(tick_pos)
    ax.set_xticklabels(tick_labels, rotation=45, fontsize=7)

    ax.set_ylim(0, 1)
    ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))

    ax.set_title(f'{pn}')
    ax.spines[['top','right']].set_visible(False)

    if ax == axes.flat[0]:
        ax.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick


month_map = {
    'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4,
    'May':5, 'Jun':6, 'Jul':7, 'Aug':8,
    'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12
}

df['Month'] = df['Month'].astype(str).str.strip().str[:3].str.title()
df['Month_num'] = df['Month'].map(month_map)

df['Date'] = pd.to_datetime(
    dict(year=df['Year'], month=df['Month_num'], day=df['Day']),
    errors='coerce'
)


df['Year_only'] = df['Date'].dt.year

PN_LIST = df['ZF_PN'].unique()[:6]


fig, axes = plt.subplots(3, 2, figsize=(16, 14))
fig.suptitle('년도별 주 단위 비율 (FIRM vs FORECAST)', fontsize=13, fontweight='bold')

for ax, pn in zip(axes.flat, PN_LIST):
    g = df[df['ZF_PN'] == pn].copy()


    g['Week'] = g['Date'].dt.to_period('W-MON').apply(lambda r: r.start_time)


    weekly = (g.groupby(['Year_only', 'Week', 'Firm/Forecast'])['Quantity']
               .sum().unstack(fill_value=0).reset_index())


    weekly['Total'] = weekly.sum(axis=1, numeric_only=True)
    weekly['FIRM_ratio'] = weekly.get('FIRM', 0) / weekly['Total']
    weekly['FC_ratio']   = weekly.get('FORECAST', 0) / weekly['Total']


    for year in sorted(weekly['Year_only'].dropna().unique()):
        sub = weekly[weekly['Year_only'] == year].sort_values('Week')

        x = range(len(sub))

        ax.plot(x, sub['FIRM_ratio'], label=f'FIRM {year}')
        ax.plot(x, sub['FC_ratio'], linestyle='--', label=f'FORECAST {year}')


    step = max(1, len(sub) // 10)
    tick_pos = list(range(0, len(sub), step))
    tick_labels = [str(d.date())[5:] for d in sub['Week'].iloc[tick_pos]]

    ax.set_xticks(tick_pos)
    ax.set_xticklabels(tick_labels, rotation=45, fontsize=7)

    ax.set_ylim(0, 1)
    ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))

    ax.set_title(f'{pn}')
    ax.spines[['top','right']].set_visible(False)

    if ax == axes.flat[0]:
        ax.legend(fontsize=8)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import matplotlib.dates as mdates


TARGET_YEAR = 2024

month_map = {
    'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4,
    'May':5, 'Jun':6, 'Jul':7, 'Aug':8,
    'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12
}

df['Month'] = df['Month'].astype(str).str.strip().str[:3].str.title()
df['Month_num'] = df['Month'].map(month_map)


df['Date'] = pd.to_datetime(
    dict(year=df['Year'], month=df['Month_num'], day=df['Day']),
    errors='coerce'
)


PN_LIST = df['ZF_PN'].unique()[:6]


fig, axes = plt.subplots(3, 2, figsize=(16, 14))
fig.suptitle(f'{TARGET_YEAR}년 주 단위 비율 (FIRM vs FORECAST)', fontsize=13, fontweight='bold')

for ax, pn in zip(axes.flat, PN_LIST):

    g = df[(df['ZF_PN'] == pn) & (df['Date'].dt.year == TARGET_YEAR)].copy()

    g['Week'] = g['Date'].dt.to_period('W-MON').apply(lambda r: r.start_time)

    weekly = (g.groupby(['Week', 'Firm/Forecast'])['Quantity']
               .sum().unstack(fill_value=0).sort_index())

    if len(weekly) == 0:
        continue

    weekly['Total'] = weekly.sum(axis=1)
    weekly['FIRM_ratio'] = weekly.get('FIRM', 0) / weekly['Total']
    weekly['FC_ratio']   = weekly.get('FORECAST', 0) / weekly['Total']


    x = weekly.index


    ax.fill_between(x, 0, weekly['FC_ratio'], alpha=0.3, label='FORECAST')
    ax.fill_between(x, weekly['FC_ratio'], weekly['FC_ratio'] + weekly['FIRM_ratio'],
                    alpha=0.4, label='FIRM')


    ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
    ax.tick_params(axis='x', rotation=45, labelsize=7)


    ax.set_ylim(0, 1)
    ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))

    ax.set_title(f'{pn}')
    ax.spines[['top','right']].set_visible(False)

    if ax == axes.flat[0]:
        ax.legend()

plt.tight_layout()
plt.show()
```
